```python
# count vowels in a string
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

count_vowels("Accenture")
```

```python
input="madam"
def is_palindrome(s):
    return s == s[::-1]

is_palindrome(input)
```
