```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# 1. Basic Linear Regression + Residual Error

# Data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])
# y = np.array([2, 4, 6, 8, 10])


# Create a linear regression model
model = LinearRegression()
model.fit(x, y)

# Predict the values
y_pred = model.predict(x)
print("Predicted values:", y_pred)


# Calculate residuals
residuals = y - y_pred
print("Residuals:", residuals)



```

```python
# 2. Residual Analysis

# Plot residuals
plt.scatter(y_pred, residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
# MSE (Mean Squared Error)
# MSE is a measure of the average squared difference between the predicted and actual values. 
# It gives an idea of how well the model is performing. A lower MSE indicates a better fit.
mse = np.mean(residuals**2)
print("MSE:", mse)
# Using sklearn
from sklearn.metrics import mean_squared_error
mse_sklearn = mean_squared_error(y, y_pred)
print("MSE (sklearn):", mse_sklearn)
```

```python
# 3. R-squared (Coefficient of Determination)
# R-squared is a statistical measure that represents the proportion of the variance for a dependent variable

from sklearn.metrics import r2_score
r2 = r2_score(y, y_pred)
print("R-squared:", r2)
# Meaning of R-squared:
# R-squared values range from 0 to 1. A value of 0 indicates that the model does not explain any of the variability of the response data around its mean, while a value of 1 indicates that the model explains all the variability of the response data around its mean. In general, a higher R-squared value indicates a better fit of the model to the data.
```

```python
# Underfitting
# Underfitting occurs when a model is too simple to capture the underlying patterns in the data

plt.scatter(x, y, color='blue', label='Actual Data')
# Plotting the regression line
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel("X")
plt.show()
```

```python
# Overfitting
# Overfitting occurs when a model is too complex and captures the noise in the data rather than the underlying pattern. 
# This can lead to poor generalization to new data.

from sklearn.preprocessing import PolynomialFeatures
# Create polynomial features
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
model_poly = LinearRegression()
model_poly.fit(x_poly, y)
```
