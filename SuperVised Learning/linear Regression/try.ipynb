```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
```

```python
# Get the Data
data= fetch_california_housing()
# print("===Data===",data)
df =pd.DataFrame(data.data,columns=data.feature_names)
df['price']=data.target
df.head()
```

```python
df.describe()
```

```python
# Data Validation and cleaning
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X=df.drop('price',axis=1)
Y=df['price']
X_train ,x_test,Y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape)
print("== X_train ==",X_train)
print("== x_test ==",x_test)
print("== Y_train ==",Y_train)
print("== y_test ==",y_test)

# X_train= df['MedInc'].values.reshape(-1,1)

# Y_train=df['price']
# print("===X_train===",X_train)

model=LinearRegression()
model.fit(X_train,Y_train)




```

```python

# print(X_train)
pred = model.predict(X_train)

print("Model Predicted Price:",pred)

```

```python
train_score = model.score(X_train, Y_train)
test_score = model.score(x_test, y_test)

print("Train R2:", train_score)
print("Test R2 :", test_score)
# x_test.values[0]
# type(x_test)
```

```python
from sklearn.metrics import r2_score
Y_train
print("R2 Score",r2_score(Y_train,pred))
```

StandardScaler is a preprocessing technique used in machine learning to standardize the features of a dataset. It transforms the data such that the mean of each feature becomes 0 and the standard deviation becomes 1. This is particularly useful when features have different scales, as it ensures that each feature contributes equally to the model's performance.

```python
# Improvement
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
scaler = StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(x_test)

model.fit(X_train_scaled,Y_train)

p=model.predict(X_train_scaled)
print("Model Predicted Price:",p)
print("R2 Score",r2_score(Y_train,p))
print("===X_train_scaled===",X_train_scaled)
print("===X_train ===",X_train)


# Note 
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train)
pred_rf = rf_model.predict(X_train)
print("Random Forest Model Predicted Price:", pred_rf)
print("Random Forest R2 Score:", r2_score(Y_train, pred_rf))

pred_rf_test = rf_model.predict([[8.3252,41.0,6.984127,1.023810,322.0,2.555556,37.88,-122.23]])
print("Random Forest Model Predicted Price for test input:", pred_rf_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_model.fit(X_train, Y_train)
pred_gb = gb_model.predict(X_train)
print("Gradient Boosting Model Predicted Price:", pred_gb)
print("Gradient Boosting R2 Score:", r2_score(Y_train, pred_gb))
pred_gb_test = gb_model.predict([[8.3252,41.0,6.984127,1.023810,322.0,2.555556,37.88,-122.23]])
print("Gradient Boosting Model Predicted Price for test input:", pred_gb_test)
```
