y= mx + c 

Where:-
y is the dependent variable (the value we want to predict)
m is the slope of the line (the rate of change of y with respect to x)
x is the independent variable (the value we use to make predictions)
c is the y-intercept (the value of y when x is 0)   

m can be calculated using the formula:
m = (NΣ(xy) - ΣxΣy) / (NΣ(x^2) - (Σx)^2)

Where:N is the number of data points
Σ(xy) is the sum of the product of x and y for all data points  
Σx is the sum of all x values
Σy is the sum of all y values
Σ(x^2) is the sum of the squares of x values
(Σx)^2 is the square of the sum of x values
c can be calculated using the formula:
c = (Σy - mΣx) / N
Where:N is the number of data points
Σy is the sum of all y values
m is the slope of the line (calculated using the previous formula)
Σx is the sum of all x values
Once we have the values of m and c, we can use the equation of the line to make predictions for y based on new values of x.For example, if we have a new value of x, we can calculate the predicted value of y using the formula:
y = mx + c

Simple Definition of Linear Regression:

Linear regression is a supervised machine Learning algorithm used for predicting a continuous target variable
based on one or more independent variables, the goal is  to find the best-fitting straight line that minimizes the difference between the predicted values and the actual values in the training data. The line is represented by the equation y = mx + c, where m is the slope of the line and c is the y-intercept. Linear regression can be used for both simple linear regression (with one independent variable) and multiple linear regression (with multiple independent variables).

```python
import numpy as np
```

```python

# Sample data
x = np.array([1000, 1500, 2000])  # Size (sq ft)
y = np.array([200000, 300000, 400000])  # Price
```

```python
n=len(x)
sum_x=np.sum(x)
sum_y= np.sum(y)
sum_xy = np.sum(x*y)
sum_x_squared = np.sum(x ** 2)
print(sum_x)
print(sum_y)
```

```python
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)

c = (sum_y- (m * sum_x)) / n
```

```python
def predict(x):
    return m*x+c

print(f"Prediction for 2500 will be {predict(2600)}")
```
