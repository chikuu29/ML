```python

from sklearn.tree import DecisionTreeClassifier
```

```python
x= [[25,50000],[35,60000],[45,80000],[20,20000],[30,40000]]
y= [0,1,1,0,1]

model= DecisionTreeClassifier(
    criterion='gini',
    max_depth=3,
)
model.fit(x, y)
prediction= model.predict([[28, 45000]])
print(prediction)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plot_tree(model)

plt.show()
```
